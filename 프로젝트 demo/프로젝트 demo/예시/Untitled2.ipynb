```python
y = new_ins["charges"]
x = new_ins.drop(["charges"], axis=1)
```

```python
x.head(5)
```

```python
y.head(5)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y) 
```

```python
x
```

```python
x_train
```

```python
x_test
```

```python
lr = LinearRegression().fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.predict([[19, 25, 0, 0, 1, 0, 1, 1, 0, 0, 0]])
```
