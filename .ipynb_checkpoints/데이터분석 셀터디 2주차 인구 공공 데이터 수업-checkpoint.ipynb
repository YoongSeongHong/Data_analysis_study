```python
import matplotlib.pyplot as plt
import csv
f = open('population.csv')
data = csv.reader(f)
mytown = []
for row in data:
    if '월계1동' in row[0]:
        for i in range(3, 104):
            mytown.append(int(row[i]))
# 저장된 데이터를 꺾은선 그래프로 그리기
plt.figure(dpi = 300)
# plt.style.use('ggplot') # 그리드
plt.rc('font', family = 'Malgun Gothic')
plt.title('월계1동의 인구 구조')
plt.plot(mytown)
plt.savefig('신도림.png')
plt.show()
```

```python
f = open('population_mf.csv')
data = csv.reader(f)
m = []
f = []
for row in data:
    if '제주' in row[0]:
        for i in range(3, 104):
            m.append(-int(row[i]))
            f.append(int(row[i+103]))
        break
plt.barh(range(101), m, label = '남')
plt.barh(range(101), f, label = '여')
plt.rcParams['axes.unicode_minus'] = False
plt.legend()
plt.show()
```

```python
size = [2441, 2312, 1031, 1233]
plt.rc('font', family='Malgun Gothic')
label = ['A형', 'B형', 'AB형', 'O형']
plt.axis('equal')
plt.pie(size, labels=label, autopct='%.1f%%')
plt.legend()
plt.show()
```

```python
size = [2441, 2312, 1031, 1233]
plt.rc('font', family='Malgun Gothic')
label = ['A형', 'B형', 'AB형', 'O형']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']
plt.axis('equal')
plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode=(0,0,0.1,0))
plt.legend()
plt.show()
```

```python
# 제주도의 연령대별 성별 비율을 산점도로 표현하기
import csv
import math
import matplotlib.pyplot as plt
f = open('population_mf.csv')
data = csv.reader(f)
m = []
f = []
size = []
name = input('궁금한 동네를 입력해주세요 : ')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i])+int(row[i+103])))
        break
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10,5), dpi=300)
plt.title(name+' 지역의 성별 인구 그래프')
plt.scatter(m, f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
