```python
import csv
```

```python
f = open('서울.csv')
data = csv.reader(f)
next(data)
high = [] # 최고 기온 값을 저장할 리스트 생성
low = [] # 최저 기온 값을 저장할 리스트 생성
```

```python
for row in data:
    if row[-1] != '' and row[-2] != '': # 최고 기온 값과 최저 기온 값이 존재한다면
        if 1983 <= int(row[0].split('-')[0]): # 1983년 이후 데이터라면
            if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '20': # 3월 20일 이라면
                high.append(float(row[-1])) # 최고 기온 값을 high 리스트에 저장
                low.append(float(row[-2])) # 최저 기온 값을 low 리스트에 저장
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.plot(high, 'hotpink', label='high') # high 리스트에 저장된 값을 hotpink 색으로 그리기
plt.plot(low, 'skyblue', label='low') # low 리스트에 저장된 값을 skyblue 색으로 그리기
plt.title('내 생일의 기온 변화 그래프')
plt.legend() # 범례 표시
plt.show()
```

```python
f = open('서울.csv')
data = csv.reader(f)
next(data)

# 월별 데이터를 저장할 리스트 month 생성
month = [[],[],[],[],[],[],[],[],[],[],[],[]]
for row in data:
    if row[-1] != '':
        # 월과 같은 번호의 인덱스에 월별 데이터 저장
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
plt.boxplot(month)
plt.show()
```

```python
f = open('서울.csv')
data = csv.reader(f)
next(data)
aug = []  # 8월의 최고 기온 값을 저장할 리스트
jan = []  # 1월이 최고 기온 값을 저장할 리스트

for row in data:
    month = row[0].split('-')[1] # -로 구분된 값 중 2번째 값 저장
    if row[-1] != '':
        if month == '08':  # 8월 달이라면
            aug.append(float(row[-1]))  # aug 리스트에 최고 기온 값 추가
        if month == '01':  # 1월 달이라면
            jan.append(float(row[-1]))  # jan 리스트에 최고 기온 값 추가
plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
        
```

```python
import csv
import matplotlib.pyplot as plt

f = open('서울.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))
plt.boxplot(aug)
plt.boxplot(jan)
plt.show()
```

```python
# 8월 일별 기온 데이터를 박스플롯으로 표현 (1907년~ 2023년)
import matplotlib.pyplot as plt
import csv

f = open('서울.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31):
    day.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
plt.style.use('ggplot')  # 그래프 스타일 지정
plt.figure(figsize=(10, 5), dpi=300)  # 그래프 값 생략
plt.boxplot(day, showfliers=False)  # 이상치 값 생략
plt.show()
```
