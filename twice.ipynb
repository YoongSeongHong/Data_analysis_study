# 누구 파트가 가장 많을까?

```python
import datetime

df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
df['time'] = df['end_time'] - df['start_time']

df_nayeon = df[(df['vocal1'] == 'Nayeon') | (df['vocal2'] == 'Nayeon')]
print('나연:', df_nayeon['time'].sum())
df_momo = df[(df['vocal1'] == 'Momo') | (df['vocal2'] == 'Momo')]
print('모모:', df_momo['time'].sum())
df_sana = df[(df['vocal1'] == 'Sana') | (df['vocal2'] == 'Sana')]
print('사나:', df_sana['time'].sum())
df_mina = df[(df['vocal1'] == 'Mina') | (df['vocal2'] == 'Mina')]
print('미나:', df_mina['time'].sum())
df_tzuyu = df[(df['vocal1'] == 'Tzuyu') | (df['vocal2'] == 'Tzuyu')]
print('쯔위:', df_tzuyu['time'].sum())
df_jihyo = df[(df['vocal1'] == 'Jihyo') | (df['vocal2'] == 'Jihyo')]
print('지효:', df_jihyo['time'].sum())
df_dahyun = df[(df['vocal1'] == 'Dahyun') | (df['vocal2'] == 'Dahyun')]
print('다현:', df_dahyun['time'].sum())
df_chaeyoung = df[(df['vocal1'] == 'Chaeyoung') | (df['vocal2'] == 'Chaeyoung')]
print('채영', df_chaeyoung['time'].sum())
df_jeongyeon = df[(df['vocal1'] == 'Jeongyeon') | (df['vocal2'] == 'Jeongyeon')]
print('정연', df_jeongyeon['time'].sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(df, x='song_name', multiple='stack', hue='vocal1', discrete=False)
plt.xticks(rotation = 90)
```

```python
songs = ['LikeOohAhh', 'TT', 'CheerUp', 'HeartShaker', 'KnockKnock', 'Likey', 'Signal', 'DanceTheNightAway', 'WhatIsLove', 'YesOrYes', 'Fancy', 'FeelSpecial', 'More&More', 'CryForMe', 'ICantStopMe', 'AlcoholFree', 'Scientist', 'TalkThatTalk']
nayeon_sum = [0] * 18
for i in range(18):
    df_songs = df_nayeon[df_nayeon['song_name'] == songs[i]]
    nayeon_sum[i] = df_songs['time'].sum()
    
nayeon_sum = pd.DataFrame(nayeon_sum)
nayeon_sum[0] = nayeon_sum[0].dt.seconds
nayeon_sum
```

```python
jihyo_sum = [0]*18
for i in range(18):
    df_songs = df_jihyo[df_jihyo['song_name'] == songs[i]]
    jihyo_sum[i] = df_songs['time'].sum()
    
jihyo_sum = pd.DataFrame(jihyo_sum)
jihyo_sum[0] = jihyo_sum[0].dt.seconds
jihyo_sum
```

```python
plt.rc('font', family='Malgun Gothic')
plt.bar(songs, nayeon_sum[0], alpha=0.5, label='나연')
plt.bar(songs, jihyo_sum[0], alpha=0.5, label='지효')
plt.title('나연과 지효 비교')
plt.legend()
plt.xticks(rotation=90)
```

```python
jihyo_sum2 = [0]*18
for i in range(18):
    df_songs = df_jihyo[df_jihyo['song_name'] == songs[i]]
    jihyo_sum2[i] = df_songs['time'].sum()
    
jihyo_sum2 = pd.DataFrame(jihyo_sum2)
jihyo_sum2[0] = -jihyo_sum2[0].dt.seconds

plt.rc('font', family='Malgun Gothic')
plt.barh(songs, nayeon_sum[0], alpha=0.5, label='나연')
plt.barh(songs, jihyo_sum2[0], alpha=0.5, label='지효')
plt.title('나연과 지효 비교')
plt.legend()
plt.xticks(rotation=90)
```
