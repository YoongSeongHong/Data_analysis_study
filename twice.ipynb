# 누구 파트가 가장 많을까?

```python
import datetime

df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
df['time'] = df['end_time'] - df['start_time']

df_nayeon = df[(df['vocal1'] == 'Nayeon') | (df['vocal2'] == 'Nayeon')]
print('나연:', df_nayeon['time'].sum())
df_momo = df[(df['vocal1'] == 'Momo') | (df['vocal2'] == 'Momo')]
print('모모:', df_momo['time'].sum())
df_sana = df[(df['vocal1'] == 'Sana') | (df['vocal2'] == 'Sana')]
print('사나:', df_sana['time'].sum())
df_mina = df[(df['vocal1'] == 'Mina') | (df['vocal2'] == 'Mina')]
print('미나:', df_mina['time'].sum())
df_tzuyu = df[(df['vocal1'] == 'Tzuyu') | (df['vocal2'] == 'Tzuyu')]
print('쯔위:', df_tzuyu['time'].sum())
df_jihyo = df[(df['vocal1'] == 'Jihyo') | (df['vocal2'] == 'Jihyo')]
print('지효:', df_jihyo['time'].sum())
df_dahyun = df[(df['vocal1'] == 'Dahyun') | (df['vocal2'] == 'Dahyun')]
print('다현:', df_dahyun['time'].sum())
df_chaeyoung = df[(df['vocal1'] == 'Chaeyoung') | (df['vocal2'] == 'Chaeyoung')]
print('채영', df_chaeyoung['time'].sum())
df_jeongyeon = df[(df['vocal1'] == 'Jeongyeon') | (df['vocal2'] == 'Jeongyeon')]
print('정연', df_jeongyeon['time'].sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(df, x='song_name', multiple='stack', hue='vocal1', discrete=False)
plt.xticks(rotation = 90)
```

```python
songs = ['LikeOohAhh', 'TT', 'CheerUp', 'HeartShaker', 'KnockKnock', 'Likey', 'Signal', 'DanceTheNightAway', 'WhatIsLove', 'YesOrYes', 'Fancy', 'FeelSpecial', 'More&More', 'CryForMe', 'ICantStopMe', 'AlcoholFree', 'Scientist', 'TalkThatTalk']
nayeon_sum = [0] * 18
for i in range(18):
    df_songs = df_nayeon[df_nayeon['song_name'] == songs[i]]
    nayeon_sum[i] = df_songs['time'].sum()
    
nayeon_sum = pd.DataFrame(nayeon_sum)
nayeon_sum[0] = nayeon_sum[0].dt.seconds
nayeon_sum
```

```python
jihyo_sum = [0]*18
for i in range(18):
    df_songs = df_jihyo[df_jihyo['song_name'] == songs[i]]
    jihyo_sum[i] = df_songs['time'].sum()
    
jihyo_sum = pd.DataFrame(jihyo_sum)
jihyo_sum[0] = jihyo_sum[0].dt.seconds
jihyo_sum
```

```python
plt.rc('font', family='Malgun Gothic')
plt.bar(songs, nayeon_sum[0], alpha=0.5, label='나연')
plt.bar(songs, jihyo_sum[0], alpha=0.5, label='지효')
plt.title('나연과 지효 비교')
plt.legend()
plt.xticks(rotation=90)
```

```python
jihyo_sum2 = [0]*18
for i in range(18):
    df_songs = df_jihyo[df_jihyo['song_name'] == songs[i]]
    jihyo_sum2[i] = df_songs['time'].sum()
    
jihyo_sum2 = pd.DataFrame(jihyo_sum2)
jihyo_sum2[0] = -jihyo_sum2[0].dt.seconds

plt.rc('font', family='Malgun Gothic')
plt.barh(songs, nayeon_sum[0], alpha=0.5, label='나연')
plt.barh(songs, jihyo_sum2[0], alpha=0.5, label='지효')
plt.title('나연과 지효 비교')
plt.legend()
plt.xticks(rotation=90)
```

# 다른 사람이 작성한 코드(멤버별 파트 시각화)

```python
import numpy as np
import pandas as pd

twice_data = pd.read_csv('twice.csv', encoding='cp949')
```

```python
twice_data.shape  # 1068개의 행, 8개의 열
```

```python
twice_data.head()
```

```python
# 위에서 나는 노래 제목들을 일일이 배열에 저장해서 썼었는데, unique를 쓰면 훨씬 편하다.
twice_data.song_name.unique()  
```

```python
# V1에는 vocal1을 기준으로 묶고 duration의 합을 저장하며, 인덱스를 초기화하고 duration을 기준으로 내림차순
V1 = twice_data.groupby('vocal1').sum()['duration'].reset_index().sort_values('duration', ascending=False)
V1
```

```python
# V2에는 vocal2을 기준으로 묶고 duration의 합을 저장하며, 인덱스를 초기화하고 duration을 기준으로 내림차순
V2 = twice_data.groupby('vocal2').sum()['duration'].reset_index().sort_values('duration', ascending=False)
V2
```

```python
# V1, V2를 merge - outer조인, 한쪽에라도 데이터 없는 경우 NaN으로 저장
V = pd.merge(V1, V2, how='outer', left_on='vocal1', right_on='vocal2')
V
```

```python
V.fillna(0)
```

```python
V.vocal1[9] = 'ALL'
```

```python
V.fillna(0)
```

```python
V['duration'] = V.duration_x+V.duration_y
```

```python
V['duration'][9] = 7.260
V
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 5))
sns.barplot(x=V.vocal1, y=V.duration, data=V)
```

```python
def Plot_world(text):
    comment_words = ' '
    # 선택한 노래의 가사들만 뽑아서 소문자의 한 문자열로 저장하는 작업
    for val in text:
        # val은 선택한 노래의 한 소절씩을 배열 형태로 저장
        val = str(val)
        # for문 한 번 당 한 소절을 토큰화해서 배열에 저장
        tokens = val.split()
        for i in range(len(tokens)):
            # 소문자로 저장
            tokens[i] = tokens[i].lower()
            
        for words in tokens:
            comment_words = comment_words + words + ' '

    # 워드클라우드 생성
    wordcloud = WordCloud(width=1000, height=500, background_color='white', min_font_size=10).generate(comment_words)
    
    plt.figure(figsize=(12,12), facecolor='k', edgecolor='k')
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
```

```python
from wordcloud import WordCloud
s = input("노래 제목을 입력하세요 : ")
text = twice_data.lyric[twice_data.song_name == s].values
Plot_world(text)
```
