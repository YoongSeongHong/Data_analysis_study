```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('서울.csv')
```

```python
data = csv.reader(f)
```

```python
next(data)
mx_temp = []
# for row in data:
#     if row[-1] != '':
#         mx_temp.append(float(row[-1]))
# print(mx_temp)


for row in data:
    if row[-1] != '' and row[0].split('-')[1] == '01':
        mx_temp.append(float(row[-1]))
plt.plot(mx_temp, color = 'hotpink')
plt.show()
```

```python
plt.hist(mx_temp, bins=100)
plt.show()
```
