```python
import numpy as np
print(np.arange(3)) # arnage() 함수에 1개 값 입력
print(np.arange(3,7)) # arange() 함수에 2개 값 입력
print(np.arange(3,7,2)) # arange() 함수에 3개 값 입력
a = np.arange(1, 2, 0.1) # 1이상 2미만 구간에서 0.1 간격으로 실수 생성 (2 포함 x)
b = np.linspace(1, 2, 11) # 1부터 2까지 11개 구간으로 나눈 실수 생성 (2 포함 o)
print(a)
print(b)
```

```python
import numpy as np
a = np.arange(-5, 5)
print(a)
print(a<0)
print(a[a<0])
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0
print(a[mask1]) # a에서 절댓값 3보다 큰 값들만
print(a[mask2]) # a에서 절댓값이 2로 나눈 나머지가 0인 값들만
print(a[mask1+mask2]) # a에서 절댓값이 3보다 크거나 2로 나눈 나머지가 0인 값들만
print(a[mask1*mask2]) # a에서 절댓값이 3보다 크고 2로 나눈 나머지가 0인 값들만
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
mask1 = abs(x) > 50 # x에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
mask2 = abs(y) > 50 # y에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
x = x[mask1+mask2] # mask1, mask2 중 하나라도 만족하는 값 저장
y = y[mask1+mask2]
size = np.random.random(len(x)) * 100 # 크기가 x 데이터 개수 만큼인 데이터 생성
plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8) # 날짜 형식으로 1월1일부터 8일까지 인덱스 생성
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC')) # 3개의 행을 A,B,C로 가짐
df['D'] = df['A'] / df['B']
df['E'] = np.sum(df, axis=1)
df = df.sub(df['A'], axis=0)
df = df.div(df['C'], axis=0) # C열 데이터를 기준으로 열 우선(axis = 0) 계산
df.head() # 위에서부터 5개만 추출
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
df = pd.read_csv('population.csv', encoding='cp949', index_col=0)
df = df.div(df['총인구수'], axis=0)
del df['총인구수'], df['연령구간인구수']

name = input('원하는 지역의 이름을 입력해주세요 : ')
a = df.index.str.contains(name)
df2 = df[a]

df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_values().index[:5]].T.plot()
plt.show()
```
